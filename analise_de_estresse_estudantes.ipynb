```python
df_para_plot = pd.melt(df_completo,
                       id_vars=['nivel_estresse'],
                       value_vars=['grupo_bullying', 'grupo_ansiedade', 'grupo_depressao'],
                       var_name='fator_de_risco',
                       value_name='grupo')

df_para_plot['fator_de_risco'] = df_para_plot['fator_de_risco'].replace({
    'grupo_bullying': 'Bullying',
    'grupo_ansiedade': 'Ansiedade',
    'grupo_depressao': 'DepressÃ£o'
})

print(df_para_plot.head())
```

```python
ordem_completa = ['Baixo','MÃ©dio', 'Alto']

plt.figure(figsize=(12, 8))


sns.pointplot(
    data=df_para_plot,
    x='grupo',
    y='nivel_estresse',
    hue='fator_de_risco',
    order=ordem_completa,
    dodge=True
)

plt.title('NÃ­vel de Estresse por Fator de Risco', fontsize=16)
plt.xlabel('NÃ­vel de ExposiÃ§Ã£o ao Fator', fontsize=12)
plt.ylabel('NÃ­vel MÃ©dio de Estresse', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Fator de Risco')
plt.tight_layout()
plt.show()
```
